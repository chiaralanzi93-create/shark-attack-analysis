```python
import pandas as pd
import numpy as np

df = pd.read_excel("https://www.sharkattackfile.net/spreadsheets/GSAF5.xls")


# valid month names
valid_months = [
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
]

def extract_month(date_str):
    if pd.isna(date_str) or not isinstance(date_str, str):  #check digit value and NaN
        return np.nan
    for month in valid_months:
        if month in date_str:
            return month
    return np.nan

# Apply the extract_month function to the 'Date' column
df['Month'] = df['Date'].apply(extract_month)

# Print the first few rows of the DataFrame to see the changes
display(df['Month'].head())
display(df['Month'].tail())
```

```python
# Compile the column Year : transfer the year from column Date into Year if the value is missing in Year column
# Function to extract year from Date column

import pandas as pd
import re

df = pd.read_excel("https://www.sharkattackfile.net/spreadsheets/GSAF5.xls")

def extract_year_from_date(date_str):
# Use regex to look for a 4-digit year. This pattern matches a sequence of exactly four digits surrounded by word boundaries,

    match = re.search(r'\b\d{4}\b', str(date_str))

    if match:
# re.match & group work together --> group() pulls out the specific part that matched.

        return int(match.group())

    return np.nan

# Transfer year from Date to Year columns if Year is NaN

df['Year'] = df.apply(lambda row: row['Year'] if not pd.isna(row['Year']) else extract_year_from_date(row['Date']),axis=1)

# Print the DataFrame with the newly extracted Year column
print(df)
```

```python
## Convert to integer, handling NaN values

df['Year'] = df['Year'].fillna(0).astype(int)
print (df)
```

```python
import re

def clean_activity(activity):
    # Check if the entry is a string
    if isinstance(activity, str):
        # Use regular expression to find the first word before any delimiters
        match = re.match(r'^([^\s\/;\-\.,]+)', activity)
        if match:
            return match.group(1)
        else:
            return activity.strip()
    else:
        return ''  # or your desired default for non-string entries

# Apply the function to the DataFrame
df['clean_activity'] = df['Activity'].apply(clean_activity)

print(df['clean_activity'].head(400))  # Print the first few entries of the new column
```

```python
#made a backup of the original column#
df["Age_raw"] = df["Age"]
```

```python
#to refine the data frame --> convert everything to string/ lowercasing everything /removing spaces at the beginning and end
df["Age"] = df["Age"].astype(str).str.lower().str.strip()
```

```python
#This checks weather the value in the Age column and contain a digit (0–9)?” and look for no digit values
no_digits = ~df["Age"].str.contains(r"\d", na=False)
df.loc[no_digits, "Age"].unique()
```

```python
# Fix duplicates using replace
df["Age"] = df["Age"].replace({
    "(adult)": "adult",
    '"young"': "young",
    '"middle-age"': "middle age",
    "teens": "teen",
})
```

```python
# We used the pandas .isin() method to find all junk age values (like "?", "m", "!!", "x", etc.)
#  and replaced them with pd.NA, so the Age column contains only meaningful information. -->  removed junk values
junk_values = ["?", "m", "!!", "", "x", "make line green", "f", "a.m."]

df.loc[df["Age"].isin(junk_values), "Age"] = pd.NA
```

```python
# Extract the first number inside each Age value
df["Age_num"] = df["Age"].str.extract(r"(\d+)")
```

```python
# Convert extracted strings into real numbers #float/integer
df["Age_num"] = pd.to_numeric(df["Age_num"], errors="coerce")
# errors="coerce" means “If something cannot be converted (like a word), turn it into NaN.”
```

```python
#check:
df[["Age", "Age_num"]].head(20)
```

```python
# to give the emaningful age groups a numerical age
age_map = {
    "a minor": 1,
    "teen": 15,
    "young": 20,
    "adult": 30,
    "middle age": 45,
    "elderly": 70
}

df["Age_num"] = df["Age_num"].fillna(df["Age"].map(age_map))
```

```python
#to check:
df.loc[df["Age"].isin(age_map.keys()), ["Age", "Age_num"]].head(20)
```

```python
import pandas as pd
import numpy as np

conditions = [
    df["Age_num"] < 13,                     # a minor
    df["Age_num"].between(13, 19, inclusive="both"),  # teen
    df["Age_num"].between(20, 29, inclusive="both"),  # young
    df["Age_num"].between(30, 39, inclusive="both"),  # adult
    df["Age_num"].between(40, 64, inclusive="both"),  # middle age
    df["Age_num"] >= 65                     # elderly
]

choices = ["a minor", "teen", "young", "adult", "middle age", "elderly"]

# 1) use a string as default so all outputs are strings
df["age_group"] = np.select(conditions, choices, default="unknown")

# 2) optional: turn "unknown" into proper missing values
df["age_group"] = df["age_group"].replace("unknown", pd.NA)
```
